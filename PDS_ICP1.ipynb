import libraries such as numpy and pandas

```python
import pandas as pd
import numpy as np
```

Loading the dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
df.head(10)
```

displaying first 15rows


```python
df.head(15)
```

Getting all numeric values columns

```python
numeric_df = df.select_dtypes(include=np.number).copy()
numeric_df
```

Checking for missing values in numeric data set

```python
numeric_df.isnull().sum()
```

Calculation mean for each column

```python
numeric_df.mean()
```

Replacing missing values with mean for numeric dataframe

```python
numeric_df.fillna(numeric_df.mean(), inplace=True)
numeric_df.isnull().sum()
```
